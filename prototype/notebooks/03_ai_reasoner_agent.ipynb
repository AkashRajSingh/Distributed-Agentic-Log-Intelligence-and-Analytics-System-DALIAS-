```python
import random
import json
from sklearn.metrics import accuracy_score

# Unified API wrapper
def call_ai_api(payload):
    # Simulated AI model (replace later with real API)
    msg = payload["log"]
    if "ERROR" in msg or "Failure" in msg:
        return {"cause": "system_error"}
    return {"cause": "normal_op"}

# Create tiny labeled dataset
def make_dataset(n=200):
    data = []
    labels = []
    for _ in range(n):
        is_err = random.random() < 0.3
        log = json.dumps({
            "level": "ERROR" if is_err else "INFO",
            "message": "Disk Failure" if is_err else "OK"
        })
        data.append(log)
        labels.append("system_error" if is_err else "normal_op")
    return data, labels

logs, labels = make_dataset()

preds = []
for l in logs:
    out = call_ai_api({"log": l})
    preds.append(out["cause"])

accuracy = accuracy_score(labels, preds)
print("Reasoner Accuracy:", accuracy)
```
